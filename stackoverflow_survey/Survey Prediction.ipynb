```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('./Data/survey_results_public.csv')
```

```python
df.head()
```

```python
df = df[['Country', 'EdLevel', 'YearsCodePro', 'Employment', 'ConvertedComp']]
df = df.rename({'ConvertedComp': 'Salary'}, axis=1)
df.head()
```

```python
df = df[df['Salary'].notnull()]
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df = df[df['Employment'] == 'Employed full-time']
df = df.drop('Employment', axis=1)
df.info()
```

```python
df['Country'].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map = {}
    for i in range(len(categories)):
        if categories.values[i] > cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(df.Country.value_counts(), 400)
df['Country'] = df['Country'].map(country_map)
df.Country.value_counts()
```

```python
f, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) Vs Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df['Salary'] <= 250000]
df = df[df['Salary'] >= 10000]
df = df[df['Country'] != 'Other']
```

```python
f, ax = plt.subplots(1, 1, figsize=(12, 7))
df.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) Vs Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=90)
plt.show()
```

```python
df['YearsCodePro'].unique()
```

```python
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    else:
        return float(x)
    
df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(x):
    if "Bachelor’s degree" in x:
        return "Bachelor’s degree"
    if "Master’s degree" in x:
        return "Master’s degree"
    if "Professional degree" in x or 'Other doctoral' in x:
        return 'Post grad'
    else:
        return 'Less than a Bachelors'
    
df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df['EdLevel'].unique()
```

```python
education_encoder = LabelEncoder()
df['EdLevel'] = education_encoder.fit_transform(df['EdLevel'])
df['EdLevel'].unique()
```

```python
country_encoder = LabelEncoder()
df['Country'] = country_encoder.fit_transform(df['Country'])
df['Country'].unique()
```

```python
X = df.drop('Salary', axis=1)
y = df['Salary']
```

```python
model = LinearRegression()
model.fit(X, y.values)
```

```python
prediction = model.predict(X)
```

```python
print(np.sqrt(mean_squared_error(y, prediction)))
```

```python
model = DecisionTreeRegressor(random_state=101)
model.fit(X, y.values)
```

```python
prediction = model.predict(X)
```

```python
print(np.sqrt(mean_squared_error(y, prediction)))
```

```python
model = RandomForestRegressor(random_state=101)
model.fit(X, y.values)
```

```python
prediction = model.predict(X)
```

```python
print(np.sqrt(mean_squared_error(y, prediction)))
```

```python
max_depth = [None, 0.1, 0.5, 1, 2, 5, 10]
parameters = {"max_depth": max_depth}

model = DecisionTreeRegressor(random_state=101)
grid = GridSearchCV(model, parameters, scoring='neg_mean_squared_error')
grid.fit(X, y.values)
```

```python
model = grid.best_estimator_
model.fit(X, y.values)
```

```python
prediction = model.predict(X)
```

```python
print(np.sqrt(mean_squared_error(y, prediction)))
```

```python
data = {'model': model, 'country_encoder': country_encoder, 'education_encoder': education_encoder}
with open('saved_steps.pkl', 'wb') as file:
    pickle.dump(data, file)
```
